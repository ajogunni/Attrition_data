```python
df = pd.DataFrame(data)
```

```python
df = pd.get_dummies(df, columns=['HallwayType', 'HeatingType','AptManageType','TimeToBusStop','TimeToSubway','SubwayStation'])
```

```python
df.info()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('SalePrice($)', axis=1), df['SalePrice($)'], test_size=0.2, random_state=42)
```

```python
# Fit a multiple linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Make predictions on test data
y_pred = regressor.predict(X_test)

print("Intercept: \n",regressor.intercept_)
print("Coefficients: \n",regressor.coef_)
```

```python
print('R-squared: {:.3f}'.format(r2_score(y_test, y_pred)))
print('MSE: {:.3f}'.format(mean_squared_error(y_test, y_pred)))
print('RMSE: {:.3f}'.format(sqrt(mean_squared_error(y_test, y_pred))))
print('MAE: {:.3f}'.format(mean_absolute_error(y_test, y_pred)))
```
